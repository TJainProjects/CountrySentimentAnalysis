```python
candidate_labels = ["economy", "security", "technology", "values", "politics", "health", "education"]
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
def classify_theme(text):
    result = classifier(str(text), candidate_labels)
    return result["labels"][0]
```

```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
```

```python
countries = [ "Russia"]
```

```python
for country in countries:

    file_path = f"/content/drive/MyDrive/GlobalChinaMapping/SentimentAnalysis_ChinaXCountry/China_{country}.xlsx"
    print(f"📂 Loading file for {country}: {file_path}")

    try:
        df = pd.read_excel(file_path)
        print(f"✅ {country} file loaded. Rows: {len(df)}")

        df["theme"] = df["snippet"].apply(classify_theme)

        df.to_excel(file_path, index=False)
        print(f"✅ Updated file saved for {country}.")

    except FileNotFoundError:
        print(f"❌ File not found for {country}. Skipping.")

```
