```python
import requests
import pandas as pd

API_KEY = "39cd6a2094b99245bd366ac2bc3255e0a3f02cba40c2c6705f5727f2550289c7"
query = "(China OR Chinese) (Hungary OR Hungarian)"

articles = []
results_per_page = 20
max_pages = 20
```

Getting the date into the required format

```python
def extract_date_from_url(url):
    try:
        res = requests.get(url, timeout=10)
        soup = BeautifulSoup(res.text, "html.parser")
        meta = soup.find("meta", {"property": "article:published_time"}) \
            or soup.find("meta", {"name": "date"}) \
            or soup.find("meta", {"name": "pubdate"}) \
            or soup.find("meta", {"itemprop": "datePublished"}) \
            or soup.find("time")
        if meta:
            return meta.get("content") or meta.get("datetime") or meta.text.strip()
        return ""
    except Exception as e:
        print(f"âš ï¸ Error scraping date from {url}: {e}")
        return ""
```

Finding all articles related to the required countries

```python
for page in range(max_pages):
    print(f"ðŸ” Fetching page {page + 1}")
    start = page * results_per_page

    params = {
        "engine": "google",
        "q": query,
        "api_key": API_KEY,
        "hl": "en",
        "num": results_per_page,
        "start": start,
        "tbs": "cdr:1,cd_min:1/1/2020,cd_max:12/31/2025"
    }

    response = requests.get("https://serpapi.com/search", params=params)
    data = response.json()

    for res in data.get("organic_results", []):
        articles.append({
            "title": res.get("title"),
            "link": res.get("link"),
            "snippet": res.get("snippet"),
            "date": res.get("date")
        })


    time.sleep(1)

    df = pd.DataFrame(articles)
```

Again reformatting the date as required

```python
def normalize_date_safe(raw_date):
    try:
        raw_date = str(raw_date).strip().lower()

        # Handle relative dates
        if "hour ago" in raw_date or "hours ago" in raw_date:
            hours = int(raw_date.split()[0])
            dt = datetime.now() - timedelta(hours=hours)

        elif "day ago" in raw_date or "days ago" in raw_date:
            days = int(raw_date.split()[0])
            dt = datetime.now() - timedelta(days=days)

        elif "minute ago" in raw_date or "minutes ago" in raw_date:
            minutes = int(raw_date.split()[0])
            dt = datetime.now() - timedelta(minutes=minutes)

        elif "just now" in raw_date:
            dt = datetime.now()

        else:
            dt = pd.to_datetime(raw_date, errors='raise')

        return dt.strftime("%b %d, %Y")
    except:
        return None


df["date_normalized"] = df["date"].apply(normalize_date_safe)

df_cleaned = df.dropna(subset=["date_normalized"]).copy()


df_cleaned.loc[:, "date"] = df_cleaned["date_normalized"]
df_cleaned.drop(columns=["date_normalized"], inplace=True)


```

```python
df = df_cleaned.copy()
df
```

Adding in the news press freedom according to the countruy of the publication of the mentioned article

```python
df["source"] = df["link"].apply(lambda x: urlparse(x).netloc.replace("www.", ""))

def get_country_from_domain(domain):
    domain = domain.lower()
    if ".ke" in domain:
        return "Kenya"
    elif ".fr" in domain:
      return "France"
    elif ".kh" in domain:
      return "Cambodia"
    elif ".vn" in domain:
      return "Vietnam"
    elif ".hn" in domain:
      return "Hungary"

    elif ".pl" in domain:
      return "Poland"
    elif ".af" in domain:
      return "Afghanistan"
    elif ".in" in domain:
      return "India"
    elif ".id" in domain:
      return "Indonesia"
    elif ".ae" in domain:
      return "United Arab Emirates"
    elif ".ch" in domain:
      return "Switzerland"
    elif ".lk" in domain:
      return "Sri Lanka"
    elif ".cn" in domain:
        return "China"
    elif ".mm" in domain:
        return "Myanmar"

    elif ".uk" in domain:
        return "UK"
    elif ".eg" in domain:
        return "Egypt"
    elif ".us" in domain or "thehill" in domain:
        return "USA"
    elif ".com" in domain:
        if "reuters" in domain:
            return "International"
        elif "bbc" in domain:
            return "UK"
        elif "cnn" in domain:
            return "USA"
    return "Unknown"

df["source_country"] = df["source"].apply(get_country_from_domain)

rsf_scores = {
    "Kenya": 49.6,
    'India': 31.0,
    "China": 22.7,
    "UK": 88.4,
    "USA": 71.2,
    "International": 80.0,
    "Unknown": 50.0,
    'France':	76.0,
    "Egypt":24.27,
    "Switzerland":84.81,
    "United Arab Emirates": 71.6,
    "Sri Lanka" : 49.46,
    "Poland":74.79,
    "Myanmar": 24.41,
    "Indonesia": 44.13,
    "Cambodia": 32.27,
    "Vietnam" :21.27,
    "Afghanistan":17.88,
    "Hungary": 62.17
}
df["rsf_score"] = df["source_country"].map(rsf_scores)

df["weight"] = df["rsf_score"] / 100

state_media = ["mfa.gov.cn", "xinhua.net", "cppcc.gov.cn"]
df["weight"] = df.apply(
    lambda x: x["weight"] * 0.3 if x["source"] in state_media else x["weight"],
    axis=1
)
```

Doing a sentiment anaklyis on the articles while taking the press freedom into account

```python
analyzer = SentimentIntensityAnalyzer()

def get_sentiment_label(text):
    score = analyzer.polarity_scores(str(text))["compound"]
    if score >= 0.05:
        return "Positive"
    elif score <= -0.05:
        return "Negative"
    else:
        return "Neutral"

def get_sentiment_score(text):
    return analyzer.polarity_scores(str(text))["compound"]

df["sentiment"] = df["snippet"].apply(get_sentiment_label)
df["sentiment_score"] = df["snippet"].apply(get_sentiment_score)

df["weighted_sentiment_score"] = df["sentiment_score"] * df["weight"]
```
